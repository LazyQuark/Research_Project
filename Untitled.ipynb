```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar

# --- KronigPenney parameters ---
a = 1.0        # lattice constant
b = 0.2        # barrier width
U = 60.0       # barrier height
d = a - b      # well width

# energy range to scan
E_min, E_max = 0.0, 80.0
num_E = 4000
E_scan = np.linspace(E_min, E_max, num_E)

# k range in first Brillouin zone
Nk = 200
k_vals = np.linspace(-np.pi/a, np.pi/a, Nk)

# dispersion function F(E,k) = cos(ka) - RHS
def dispersion(E, k):
    if E <= 0:
        return 10  # avoid sqrt error

    alpha = np.sqrt(2*E)
    beta = np.sqrt(2*(U - E)) if E < U else 1e-9  # evanescent

    RHS = (np.cos(alpha*d)*np.cosh(beta*b)
           + (beta**2 - alpha**2)/(2*alpha*beta)
             * np.sin(alpha*d)*np.sinh(beta*b))

    return np.cos(k*a) - RHS


# ---- Find roots for each k ----
bands = []
max_bands = 4    # how many bands to try to extract
tolerance = 0.1

for k in k_vals:
    band_E = []

    # scan for sign changes in dispersion relation
    for i in range(num_E - 1):
        f1 = dispersion(E_scan[i], k)
        f2 = dispersion(E_scan[i+1], k)

        if np.sign(f1) != np.sign(f2):
            # root bracket found
            try:
                sol = root_scalar(dispersion, args=(k,),
                                  bracket=[E_scan[i], E_scan[i+1]],
                                  method='bisect')
                band_E.append(sol.root)
            except:
                pass

    bands.append(band_E)

# Convert bands list-of-lists into proper columns
# by padding missing values with NaN
max_len = max(len(b) for b in bands)
bands_array = np.full((max_len, Nk), np.nan)

for j in range(Nk):
    for i, val in enumerate(bands[j]):
        if i < max_len:
            bands_array[i, j] = val

# ---- Plot ----
plt.figure(figsize=(8,6))

colors = ['blue','green','orange','magenta']

for i in range(min(max_len, len(colors))):
    plt.plot(k_vals, bands_array[i], label=f"Band {i+1}", linewidth=2)

plt.axvline(-np.pi/a, color='gray', linestyle='--')
plt.axvline( np.pi/a, color='gray', linestyle='--')

plt.xlabel("k")
plt.ylabel("E(k)")
plt.title("Energy Dispersion for KronigPenney Model (a=1, b=0.2, U=60)")
plt.grid(alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar

# --- KronigPenney parameters ---
a = 1.0
b = 0.1
U = 60.0
d = a - b

E_min, E_max = 0.0, 80.0
num_E = 4000
E_scan = np.linspace(E_min, E_max, num_E)

Nk = 200
k_vals = np.linspace(-np.pi/a, np.pi/a, Nk)

# Dispersion relation F(E,k) = 0
def dispersion(E, k):
    if E <= 0: return 1e6
    alpha = np.sqrt(2*E)
    beta  = np.sqrt(2*(U - E)) if E < U else 1e-6
    RHS = (np.cos(alpha*d)*np.cosh(beta*b)
           + (beta**2 - alpha**2)/(2*alpha*beta)
             * np.sin(alpha*d)*np.sinh(beta*b))
    return np.cos(k*a) - RHS

# Solve for bands
bands = []
for k in k_vals:
    roots_k = []
    for i in range(num_E - 1):
        if np.sign(dispersion(E_scan[i], k)) != np.sign(dispersion(E_scan[i+1], k)):
            try:
                sol = root_scalar(dispersion, args=(k,),
                                  bracket=[E_scan[i], E_scan[i+1]],
                                  method='bisect')
                roots_k.append(sol.root)
            except:
                pass
    bands.append(roots_k)

# Convert into regular 2D array with padding
max_bands = max(len(b) for b in bands)
bands_array = np.full((max_bands, Nk), np.nan)
for j in range(Nk):
    for i, val in enumerate(bands[j]):
        bands_array[i, j] = val

# --------------------------------------------------------
# FIND BAND GAPS
# --------------------------------------------------------
band_mins = np.nanmin(bands_array, axis=1)
band_maxs = np.nanmax(bands_array, axis=1)

gaps = []
for i in range(len(band_mins) - 1):
    gap_low  = band_maxs[i]
    gap_high = band_mins[i+1]
    if gap_high > gap_low:
        gaps.append((gap_low, gap_high))

# --------------------------------------------------------
# PLOT
# --------------------------------------------------------
plt.figure(figsize=(9,6))

colors = ['blue','green','orange','magenta','red']

# Plot first two bands thicker and labeled
for i in range(min(2, max_bands)):
    plt.plot(k_vals, bands_array[i], label=f"Band {i+1}", linewidth=3, color=colors[i])

# Plot remaining bands thin
for i in range(2, max_bands):
    plt.plot(k_vals, bands_array[i], label=f"Band {i+1}", linewidth=1.5, color=colors[i])

# Shade band gaps clearly
for (g1, g2) in gaps:
    plt.axhspan(g1, g2, color='gray', alpha=0.15, label="Band gap" if g1 == gaps[0][0] else "")

# Mark Brillouin zone boundaries
plt.axvline(-np.pi/a, color='black', linestyle='--')
plt.axvline( np.pi/a, color='black', linestyle='--')

plt.xlabel("k")
plt.ylabel("E(k)")
plt.title("KronigPenney Energy Dispersion (a=1, b=0.2, U=60)")

plt.xlim(-np.pi, np.pi)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

print(gaps)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ================ PARAMETERS ======================
a = 1.0        # lattice spacing
b = 0.1        # barrier width
U = 75.0       # barrier height

# ================ DISPERSION FUNCTION =============
def F(E):
    """ KronigPenney transcendental function f(E) = cos(ka). """
    k1 = np.sqrt(E)
    k2 = np.sqrt(E - U)

    # handle E < U by using imaginary wave number
    k2 = np.where(E >= U, k2, 1j * np.sqrt(U - E))

    term1 = np.cos(k1 * (a - b)) * np.cos(k2 * b)
    term2 = (k1**2 + k2**2) / (2 * k1 * k2) * np.sin(k1 * (a - b)) * np.sin(k2 * b)
    return np.real(term1 - term2)

# ================ ENERGY GRID =====================
E = np.linspace(0.001, 120, 9000)   # Wider range because U = 75 is large
Fvals = F(E)

# Allowed bands satisfy |F(E)| <= 1
allowed = np.abs(Fvals) <= 1

# Identify band boundaries
bands = []
in_band = False
for i in range(len(E)):
    if allowed[i] and not in_band:
        start = E[i]
        in_band = True
    if not allowed[i] and in_band:
        end = E[i-1]
        in_band = False
        bands.append((start, end))
if in_band:
    bands.append((start, E[-1]))

# Compute band gaps
gaps = []
for i in range(len(bands)-1):
    gaps.append(bands[i+1][0] - bands[i][1])

# =================== PLOT ==========================
plt.figure(figsize=(9,6))
plt.plot(E, Fvals, linewidth=1)
plt.axhline(1, color='black', linestyle='--', linewidth=0.8)
plt.axhline(-1, color='black', linestyle='--', linewidth=0.8)

# Shade allowed bands
for (E1, E2) in bands:
    idx = np.logical_and(E >= E1, E <= E2)
    plt.fill_between(E[idx], Fvals[idx], 1.1, color='lightgreen', alpha=0.4)

# Mark band edges for first two bands
for (E1, E2) in bands[:2]:
    plt.axvline(E1, color='red', linestyle=':', linewidth=1.2)
    plt.axvline(E2, color='red', linestyle=':', linewidth=1.2)

# Write gap widths on the figure
for i, g in enumerate(gaps[:3]):
    mid = (bands[i][1] + bands[i+1][0]) / 2
    plt.text(mid, 1.15, f"Gap {i+1} = {g:.2f}",
             ha='center', fontsize=10, color='purple')

plt.title(f"KronigPenney Dispersion (a={a}, b={b}, U={U})")
plt.xlabel("Energy E")
plt.ylabel("F(E)")
plt.ylim(-2, 2)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
