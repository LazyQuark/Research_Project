```python
from pycbc import noise, psd
from pycbc.psd import welch , interpolate , aLIGOZeroDetHighPower
from pycbc.waveform import get_td_waveform , get_fd_waveform
from pycbc.filter import highpass_fir , lowpass_fir
from pycbc.filter import match , matched_filter

import matplotlib.pyplot as plt
import numpy as np


# first lets pretend we have a signal in our data and lets make it like a time series data

f_low = 30.0
delta_f = 1.0 / 16
flen = int(2048 / delta_f) + 1    # 2048 is the max freq (nyquiest)
psd = aLIGOZeroDetHighPower(flen, delta_f, f_low)   # we generated a psd to mimic detector noise

# noise creation 
sps = 4096
dt = 1.0 / sps
tsamples = int(40 / dt)          #40 seconds of noise at 4096 Hz
noise = noise.noise_from_psd(tsamples, dt, psd,seed = 130)   # so we get a noise for 128seconds at sps 4096 


# the waveform we have:

hp, hc = get_td_waveform(approximant="IMRPhenomD",mass1=30,mass2=40,delta_t=dt,f_lower=f_low)  # noise and hp spacing are same
#print(hp.start_time) # -11
#print(noise.start_time) # 0


# print(hp.duration)  # 16 seconds signal 
# print(noise.duration) # 40
# now lets put this 16 seconds signal  in that 40 seconds noise at the 15 seconds time onwards

time_of_injection = 15 # seconds
hp._epoch = noise.start_time + time_of_injection
# print(hp.start_time)  # now set to 15 seconds

# lets calculate the index of this 15 seconds array

index_toi = (time_of_injection * sps)  

noise_data = noise.copy()
noise_data[index_toi: index_toi + len(hp)]  += hp.data*0.001  # we added the hp to noise data at 15 seconds and made a weak signal


##################################
## noise data plotting
#################################
plt.figure(figsize = (12,8))
plt.plot(noise_data.sample_times , noise_data , label = "noisy data")
plt.title("Noise data of 40 seconds",size = 15)
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Strain")
plt.show()
#plt.ylim([-2e-21 , 2e-21])





###############################################################
#
#                   ANALYSIS STARTS HERE
#
###############################################################
## lets see if there is signal in this noise with match filter for various templates with diff masses

noise_data = highpass_fir(noise_data , 30,8)   # filtering the low noise freq
noise_data_tilde = noise_data.to_frequencyseries()

# PSD to match data's delta_f
psd_interp = interpolate(psd, noise_data_tilde.delta_f)



m1_vals = np.arange(10, 50, 2) 
m2_vals = np.arange(10, 50, 2)   

snr_grid = np.zeros((len(m1_vals), len(m2_vals)))   # we are creating a 2d matrxi

# print(len(noise_data),len(noise_data_tilde))
#print(noise_data_tilde.delta_f)

global_snr_max = 0
best_mass = (0, 0) 


for i,m1 in enumerate(m1_vals):
    for j,m2 in enumerate(m2_vals):
        if m1 >= m2:
            
            hp_tem , hc_tem = get_fd_waveform(approximant = "IMRPhenomD" , mass1 = m1 , mass2 = m2,
                                                f_lower = f_low , delta_f = noise_data_tilde.delta_f)
            hp_tem.resize(len(noise_data_tilde))
            snr = matched_filter(hp_tem, noise_data_tilde, psd=psd_interp, low_frequency_cutoff=f_low)
            
            max_snr = abs(snr).max()
            snr_grid[i, j] = max_snr
            
            if max_snr > global_snr_max:
                global_snr_max = max_snr
                best_mass = (m1, m2)
         
        else:
            snr_grid[i,j] = None

print(f"\nSearch complete!")


# Plotting
plt.figure(figsize=(12, 9))

plt.imshow(snr_grid, 
           extent=[m1_vals[0], m1_vals[-1], 
                   m2_vals[0], m2_vals[-1]],
           cmap='plasma', origin='lower', aspect='auto')

# Mark best match and published value
plt.plot(best_mass[1], best_mass[0], 'w*', markersize=30, 
         label=f'Best match: ({best_mass[0]:.1f}, {best_mass[1]:.1f}) M', 
         markeredgecolor='black', markeredgewidth=1.5)

plt.colorbar(label='SNR')
plt.xlabel('Mass 2 (Solar Masses)', fontsize=12)
plt.ylabel('Mass 1 (Solar Masses)', fontsize=12)
plt.title(f'Matched Filter Analysis\nMax SNR: {global_snr_max:.2f}', fontsize=14)
plt.legend(fontsize=11)
plt.grid(alpha=0.3)

plt.show()

print(f"\nBest matching masses: M1={best_mass[0]:.1f} M, M2={best_mass[1]:.1f} M")
print(f"Maximum SNR: {global_snr_max:.2f}")
```
