```python
##########################################################
####### LIBRARIES 
##########################################################
from pycbc import noise, psd
from pycbc.psd import welch , interpolate , aLIGOZeroDetHighPower
from pycbc.waveform import get_td_waveform , get_fd_waveform
from pycbc.filter import highpass_fir , lowpass_fir
from pycbc.filter import match , matched_filter
import matplotlib.pyplot as plt
import numpy as np
import time
##########################################################
# Creating a data series with imbedded known signal for analysis
##########################################################

# first lets pretend we have a signal in our data and lets make it like a time series data
start = time.time()
T = 40.0
sps = 4096
f_low = 10.0    # detector range

dt = 1/sps      # time spacing
df = 1.0 / T  # freq spacing
N = int(sps*T)     # total points  in time
f_N = int(N/2) +1  # total points in freq


########################## noise creation ##############################

psd = aLIGOZeroDetHighPower(f_N, df, f_low)   # we generated a psd to mimic detector noise
noise = noise.noise_from_psd(N, dt, psd,seed = 130)   # so we get a noise for 40seconds at sps 4096 


################### the waveform we have:###############################

hp, hc = get_td_waveform(approximant="IMRPhenomD",mass1=30,mass2=40,delta_t=dt,f_lower=f_low,distance = 1000)  # noise,hp spacing 
print(hp.start_time) # -27
print(noise.start_time) # 0
print(hp.duration)

hp.resize(len(noise))
print(hp.duration)  
# print(noise.duration) # 40
############## now lets put this 16 seconds signal  in that 40 seconds noise at the 15 seconds time onwards#######

#time_of_injection =  0 # seconds
#hp._epoch = noise.start_time + time_of_injection
# print(hp.start_time)  # now set to 15 seconds

# lets calculate the index of this 15 seconds array

#index_toi = (time_of_injection * sps)  

noise_data = noise.copy()
#noise_data[index_toi: index_toi + len(hp)]  += hp.data*0.001  # we added the hp to noise data at 15 seconds and  
noise_data = noise_data+ hp.data                         # a weak signal
noise_data = highpass_fir(noise_data , 10,8)   # filtering the low noise freq

##################################################################
## noise data plotting
##################################################################
plt.figure(figsize = (12,4))
plt.plot(noise_data.sample_times , noise_data , label = "noisy data")
plt.title("Noise data of 40 seconds",size = 15)
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Strain")
plt.show()
#plt.ylim([-2e-21 , 2e-21])





#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
# till now we have a noise data that may or may not contain a signal








###############################################################
#
#                   ANALYSIS STARTS HERE
### lets see if there is signal in this noise with match filter for various templates with diff masses
###############################################################


#noise_data = highpass_fir(noise_data , 10,8)   # filtering the low noise freq
#noise_data_tilde = noise_data.to_frequencyseries()




m1_vals = np.arange(10, 50, 2) 
m2_vals = np.arange(10, 50, 2)   

snr_grid = np.zeros((len(m1_vals), len(m2_vals)))   # we are creating a 2d matrxi

# print(len(noise_data),len(noise_data_tilde))
#print(noise_data_tilde.delta_f)
##############################   INITIALIZATION   ####################################
global_snr_max = 0
best_mass = (0, 0) 
time_of_arrival = 0

for i,m1 in enumerate(m1_vals):
    for j,m2 in enumerate(m2_vals):
        if m1 >= m2:
            
            hp_tem , hc_tem = get_td_waveform(approximant = "IMRPhenomD" , mass1 = m1 , mass2 = m2,
                                                f_lower = f_low , delta_t = dt)
            
            #print(hp_tem.duration)
            hp_tem.resize(len(noise))
            snr = matched_filter(hp_tem, noise_data, psd=psd, low_frequency_cutoff=f_low)
            
            max_snr = abs(snr).max()
            snr_grid[i, j] = max_snr
            
            if max_snr > global_snr_max:
                global_snr_max = max_snr
                best_mass = (m1, m2)
                
                max_idx = abs(snr).numpy().argmax()   # index where SNR is maximum
                time_of_arrival = snr.sample_times[max_idx]  
         
        else:
            snr_grid[i,j] = None      # ignoring other symmetric part

print(f"\nSearch complete!")









######################################
######          Plotting
####################################
plt.figure(figsize=(12, 8))

plt.imshow(snr_grid, 
           extent=[m1_vals[0], m1_vals[-1], 
                   m2_vals[0], m2_vals[-1]],
           cmap='plasma', origin='lower', aspect='auto')

# Mark best match and published value
plt.plot(best_mass[1], best_mass[0], 'w*', markersize=30, 
         label=f'Best match: ({best_mass[0]:.1f}, {best_mass[1]:.1f}) M', 
         markeredgecolor='black', markeredgewidth=1.5)

plt.colorbar(label='SNR')
plt.xlabel('Mass 2 (Solar Masses)', fontsize=12)
plt.ylabel('Mass 1 (Solar Masses)', fontsize=12)
plt.title(f'Matched Filter Analysis\nMax SNR: {global_snr_max:.2f}', fontsize=14)
plt.legend(fontsize=11)
plt.axvline(best_mass[1],linestyle = "--",alpha = 0.3,color = 'black')
plt.axhline(best_mass[0],linestyle = "--",alpha = 0.3,color = 'black')
plt.grid(alpha=0.3)

plt.show()
print(f"\nBest matching masses: M1={best_mass[0]:.1f} M, M2={best_mass[1]:.1f} M")
print(f"Maximum SNR: {global_snr_max:.2f}")
print(f"Time of arrival of signal : {time_of_arrival:2.4f}")   # our time of injection



############################### BEST TEMPLATE SNR #################################
hp_tem_best, hc_tem_best = get_td_waveform(approximant = "IMRPhenomD" , mass1 = best_mass[0] , mass2 = best_mass[1],
                                                f_lower = f_low , delta_t = dt)
hp_tem_best.resize(len(noise))
snr = matched_filter(hp_tem_best, noise_data, psd=psd, low_frequency_cutoff=f_low)
plt.figure(figsize = (12,4))
plt.plot(snr.sample_times , abs(snr))
plt.legend(["SNR"])
plt.grid(alpha = 0.3)
plt.xlabel("Time [s]")
plt.ylabel("SNR")
plt.title("Best Template SNR values",size = 16)

plt.show()



end = time.time()

print(f'Time to run the code :{end-start:3.3f} seconds')

print("\n")
```
